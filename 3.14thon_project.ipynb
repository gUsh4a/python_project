# Возьмём статистику по ДТП с водителями в состоянии опьянения за сентябри 2025 и 2024 гг.

```python
sep_25_list = []
with open('Российская_Федерация_сентябрь_2025.csv') as infile:
    for line in infile:
        list_25 = line.split(';')
        sep_25_list.append(list_25)
```

Рассмотрим верхние строки:

```python
for i in range(5):
    print(sep_25_list[i])
```

Видим в 2-й строке абс и пр, абс - абслоютное значение, оставми только такие:

```python
sep_25_list = [[row[i] for i in [0, 1, 3, 5]] for row in sep_25_list]
```

```python
for i in range(5):
    print(sep_25_list[i])
```

в целом для дальнейшего удобства удалим первые 4 строки, запомнив, что первый столбец - субъект, второй - кол-во дтп, 3 - сколько погибло, 4 - сколько ранено

```python
sep_25_list = sep_25_list[4:]

sep_25_list = [row for row in sep_25_list if any(cell.strip() for cell in row)]
```

Сделаем словарь, где ключи это субъекты, округа и т.д. РФ, так как это удобнее для сравнения их по показателям ДТП


```python
sep_25_dict = {row[0]: row[1:] for row in sep_25_list}

print("Словарь:")
for key, value in sep_25_dict.items():
    print(f"  '{key}': {value}")
```

сохраним отдельно суъекты и округа

```python
okruga = {k: sep_25_dict[k] for k in sep_25_dict if 'федеральный округ' in k}

for key, value in okruga.items():
    print(f"{key}: ДТП={value[0]}, Погибло={value[1]}, Ранено={value[2]}")
```

```python
subjects = {k: v for k, v in sep_25_dict.items()
            if ('федеральный' not in k and 'территор' not in k)
            and k != 'Российская Федерация' and k != 'Сириус'}

for key, value in subjects.items():
    print(f"{key}: ДТП={value[0]}, Погибло={value[1]}, Ранено={value[2]}")
```

# СТАТИТИСТИКА

```python
max_values = {
    'ДТП': max(float(v[0]) for v in subjects.values()),
    'Погибло': max(float(v[1]) for v in subjects.values()),
    'Ранено': max(float(v[2]) for v in subjects.values())
}

print("Максимальные значения:")
for indicator, value in max_values.items():
    print(f"  {indicator}: {value}")

    max_regions = [k for k, v in subjects.items() if float(v[['ДТП', 'Погибло', 'Ранено'].index(indicator)]) == value]
    print(f"Субъект: {', '.join(max_regions[:3])}{'...' if len(max_regions) > 3 else ''}")
```

СРАВНИМ ФЕДЕРАЛЬНЫЕ ОКРУГА

```python
for district, data in sorted(okruga.items(), key=lambda x: float(x[1][0]), reverse=True):
    dtp, deaths, injured = map(float, data)
    severity = deaths / dtp * 100 if dtp > 0 else 0
    print(f"Тяжесть: {severity:.1f}% (погибших от общего числа ДТП)")
    print(f"{district}: {dtp} ДТП, {deaths} погибло, {injured} ранено")
```

```python
print("Субъекты с наименьшим кол-вом ДТП с водителями в состоянии опьянения в сентябре 2025:")
safe_regions = sorted(subjects.items(), key=lambda x: float(x[1][0]))[:10]
for i, (key, value) in enumerate(safe_regions, 1):
    print(f"{i}. {key}: {value[0]} ДТП")
```

```python
print(f"Всего субъектов: {len(subjects)}")

total_dtp = sum(float(d[0]) for d in subjects.values())
total_deaths = sum(float(d[1]) for d in subjects.values())
total_injured = sum(float(d[2]) for d in subjects.values())

print(f"Общее ДТП по регионам: {total_dtp:.0f}")
print(f"Общее погибших: {total_deaths:.0f}")
print(f"Общее раненых: {total_injured:.0f}")
```

Средние значения по субъектам

```python
mean_dtp = total_dtp / len(subjects)
mean_deaths = total_deaths / len(subjects)
mean_injured = total_injured / len(subjects)

print(f"Среднее ДТП на регион: {mean_dtp:.2f}")
print(f"Среднее погибших на регион: {mean_deaths:.2f}")
print(f"Среднее раненых на регион: {mean_injured:.2f}")

severity_list = [float(d[1])/float(d[0])*100 if float(d[0]) > 0 else 0 for d in subjects.values()]
mean_severity = sum(severity_list) / len(severity_list)
print(f"Средняя тяжесть последствий: {mean_severity:.2f}% (погибших от ДТП)")
```

Медианные значения

```python
median_dtp = sorted(float(d[0]) for d in subjects.values())[len(subjects)//2]
median_deaths = sorted(float(d[1]) for d in subjects.values())[len(subjects)//2]
median_injured = sorted(float(d[2]) for d in subjects.values())[len(subjects)//2]

print(f"Медиана ДТП по регионам: {median_dtp:.1f}")
print(f"Медиана погибших по регионам: {median_deaths:.1f}")
print(f"Медиана ранены по регионамх: {median_injured:.1f}")
```

Посчитаем остальные статистические показатели

```python
variance_dtp = sum((float(d[0]) - mean_dtp)**2 for d in subjects.values()) / len(subjects) #дисперсия
std_dev_dtp = variance_dtp ** 0.5
cv_dtp = (std_dev_dtp/mean_dtp*100)

variance_deaths = sum((float(d[1]) - mean_deaths)**2 for d in subjects.values()) / len(subjects)
std_dev_deaths = variance_deaths**0.5
cv_deaths = (std_dev_deaths/mean_deaths*100)

print(f"Стандартное отклонение по ДТП: {std_dev_dtp:.2f}")
print(f"Коэффициент вариации по ДТП: {cv_dtp:.1f}%")
print(f"Стандартное отклонение по погибшим: {std_dev_deaths:.2f}")
print(f"Коэффициент вариации по погибшим: {cv_deaths:.1f}%")
```

```python
max_dtp = max(float(d[0]) for d in subjects.values())
min_dtp = min(float(d[0]) for d in subjects.values())
max_deaths = max(float(d[1]) for d in subjects.values())
min_deaths = min(float(d[1]) for d in subjects.values())

print(f"\nДИАПАЗОНЫ ЗНАЧЕНИЙ:")
print(f"ДТП: от {min_dtp:.1f} до {max_dtp:.1f} (размах: {max_dtp - min_dtp:.1f})")
print(f"Погибшие: от {min_deaths:.1f} до {max_deaths:.1f} (размах: {max_deaths - min_deaths:.1f})")

percentiles_dtp = sorted(float(d[0]) for d in subjects.values())
p25_dtp = percentiles_dtp[int(len(percentiles_dtp)*0.25)]
p75_dtp = percentiles_dtp[int(len(percentiles_dtp)*0.75)]

print(f"Перцентили:")
print(f"25-й перцентиль ДТП: {p25_dtp:.1f}")
print(f"75-й перцентиль ДТП: {p75_dtp:.1f}")
print(f"Межквартильный размах ДТП: {p75_dtp - p25_dtp:.1f}")
```

 Сравним среднее и медиану

```python
print(f"Разница среднего и медианы ДТП: {mean_dtp - median_dtp:.2f}")
if mean_dtp > median_dtp:
    print("Среднее > медианы => распределение с правосторонней асимметрией")
elif mean_dtp < median_dtp:
    print("Среднее < медианы => распределение с левосторонней асимметрией")
else:
    print("Среднее = медиане => симметричное распределение")
```

Сделаем словарь, где ключи это субъекты, округа и т.д. РФ, так как это удобнее для сравнения их по показателям ДТП

```python
sep_24_dict = {row[0]: row[1:] for row in sep_24_list}

print("Словарь:")
for key, value in sep_24_dict.items():
    print(f"  '{key}': {value}")
```

```python
okruga_24 = {k: sep_24_dict[k] for k in sep_24_dict if 'федеральный' in k}

for key, value in okruga_24.items():
    print(f"{key}: ДТП={value[0]}, Погибло={value[1]}, Ранено={value[2]}")
```

```python
subjects_24 = {k: v for k, v in sep_24_dict.items()
            if ('федеральн' not in k and 'территор' not in k)
            and k != 'Российская Федерация' and k != 'Сириус'}

for key, value in subjects_24.items():
    print(f"{key}: ДТП={value[0]}, Погибло={value[1]}, Ранено={value[2]}")
```

```python
safe_regions_24 = sorted(subjects_24.items(), key=lambda x: float(x[1][0]))[:10]
for i, (key, value) in enumerate(safe_regions_24, 1):
    print(f"{i}. {key}: {value[0]} ДТП")
```

```python
print("ВИЗУАЛИЗАЦИЯ КОЛИЧЕСТВА ИЗМЕНЕНИЙ ДТП ПО ОКРУГАМ\n")

print(f"{'Округ':<35} {'сентябрь 2024':<20} {'сентябрь 2025':<20}")
print(f"{'':<35} {'─'*20} {'─'*20}")

max_dtp = max(max(float(d[0]) for d in okruga_24.values()),
              max(float(d[0]) for d in okruga.values()))

for district_name in sorted(okruga_24.keys()):
    if district_name in okruga:
        dtp_24_val = float(okruga_24[district_name][0])
        dtp_25_val = float(okruga[district_name][0])

        bar_24 = "█" * int(dtp_24_val / max_dtp * 20)
        bar_25 = "█" * int(dtp_25_val / max_dtp * 20)

        print(f"{district_name:<35} {bar_24:<20} {bar_25:<20}")
        print(f"{'':<35} {dtp_24_val:>5.0f}{'':<15} {dtp_25_val:>5.0f}")
```

```python
safe_regions_24 = sorted(subjects_24.items(), key=lambda x: float(x[1][0]))[:10]
safe_regions_25 = sorted(subjects.items(), key=lambda x: float(x[1][0]))[:10]

positions_24 = {region: i for i, (region, _) in enumerate(safe_regions_24, 1)}
positions_25 = {region: i for i, (region, _) in enumerate(safe_regions_25, 1)}

print(f"\n2024 год:")
for i, (region, data) in enumerate(safe_regions_24, 1):
    print(f"{i:2}. {region:<30} {float(data[0]):5.1f} ДТП")

print(f"\n2025 год:")
for i, (region, data) in enumerate(safe_regions_25, 1):
    print(f"{i:2}. {region:<30} {float(data[0]):5.1f} ДТП")
```

```python
print(f"Cтабильность субъектов с наименьшик кол-вом ДТП с водителями в состоянии опьянения:")

stayed_in_top = set([r for r, _ in safe_regions_24]) & set([r for r, _ in safe_regions_25])
entered_top = set([r for r, _ in safe_regions_25]) - set([r for r, _ in safe_regions_24])
left_top = set([r for r, _ in safe_regions_24]) - set([r for r, _ in safe_regions_25])

print(f"Осталось субъектов в ТОП-10: {len(stayed_in_top)} ")
print(f"Вошло субъектов в ТОП-10: {len(entered_top)} ")
print(f"Вышло субъектов из ТОП-10: {len(left_top)} ")

if entered_top:
    print(f"\nСубъекты, вошедшие в ТОП-10 в 2025:")
    for region in sorted(entered_top):
        dtp_25_val = float(subjects[region][0])
        print(f"  {region} - {dtp_25_val:.1f} ДТП")
```
